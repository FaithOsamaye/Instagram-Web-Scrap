```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
from os import uname_result
url='https://en.wikipedia.org/wiki/List_of_most-followed_Instagram_accounts'

page=requests.get(url)

soup=BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify)
```

```python
table = soup.find_all('table')[0] #Selecting the precise table we want.
```

```python
#Extracting the column name('th') #'th tags are our title.
columns=table.find_all('th')
print(columns)
```

```python
#List comprehension to get our unique title
column_titles= [col.text.strip() for col in columns]
column_titles = column_titles[0:6]
print(column_titles)
```

```python
instagram = pd.DataFrame(columns=column_titles)

instagram
```

```python
#The tr represents the rows, the td represents the data. td is embeded in the tr.
column_data = table.find_all('tr')
```

```python
for row in column_data[1:51]:
  row_data=row.find_all('td')
  individual_rows=[data.text.strip() for data in row_data]
  print(individual_rows)
  length=len(instagram)
  instagram.loc[length]=individual_rows

#We need to remove this first empty list.
```
